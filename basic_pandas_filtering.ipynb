**PANDAS DATAFRAME FILTERING**

Pandas DataFrame filtering is a crucial technique for selecting subsets of data based on specific conditions.


Multiple Conditions: Combine multiple conditions using logical operators (& for 'and', | for 'or').

df[(df['column1'] > value1) & (df['column2']<value2)]

These filtering techniques enable precise and efficient data manipulation, essential for data analysis and preprocessingÂ tasks.


```python
import pandas as pd
import numpy as np
```

```python
path="/content/weather_data.csv"
```

```python
df=pd.read_csv(path)
```

```python
df
```

```python
df['temperature']>30
```

```python
#find the  dataframe if temperature greaterthan 30
df[df["temperature"]>30]
```

```python
#find the  dataframe if wind-speed lessthan or equal to 6
df[df["windspeed"]<=6]
```

```python
#find the day if wind-speed less than or equal to 6?
df[df["windspeed"]<=6].day
```

```python
#find the day if wind-speed less than or equal to 6?
df['day'][df["windspeed"]<=6]
```

```python
#find the day and temperature if the wind-speed lessthan or equalto 6?
df[['day','temperature']][df['windspeed']<=6]
```

```python
#find the temperature and wind-speed if temperature lessthan or equalto 30?
df[['temperature','windspeed']][df['temperature']<=30]
```

```python
#find the day in which maximum temperature recorded?
max=df['temperature'].max()
df['day'][df['temperature']==max]
```

```python
df['day'][df['temperature']==df['temperature'].max()]
```

```python
#find the day and temperature in which minimum temperature is recorded
df['day'][df['temperature']==df['temperature'].min()]
```

```python
#find the days in which temperature wend above average?
df['day'][df['temperature']>=df["temperature"].mean()]
```

if multiple condition , each condition should be on paranthesis() ,
or == | ,
and == &

```python
#eithertemperature lessthen 30 or windspeed >5
df[(df['temperature']<30) | (df['windspeed']>5)]
```

```python
#event=snow and temperature<25
df[(df["event"]=='Snow') & (df['temperature']<25)]
```

```python
#temperature greaterthen 30 and windspeed lessthan 5 or temperature lessthan 30 and windspeed greaterthan 5
df[((df['temperature']>30) & (df['windspeed']<5)) | ((df['temperature']<30) & (df['windspeed']>5)) ]
```

***10 questions & code on data filtering using weatherÂ dataset***

```python
import pandas as pd
import numpy as np
path="/content/weather_data.csv"
df=pd.read_csv(path)
df
```

```python
#1.find dataframe where windspeed greater than 5 and event is equal to snow
df[(df['windspeed']>5) & (df['event']=='Snow')]
```

```python
#2.find dataframe where windspeed<5 and event not equal to rain?
df[(df['windspeed']<5) & (df['event']!='Rain')]
```

```python
#3.find dataframe where windspeed greater than 5 and event is equal to snow or dataframe where windspeed<5 and event not equal to rain?

df[((df['windspeed']>5) & (df['event']=='Snow')) | ((df['windspeed']<5) & (df['event']!='Rain'))]
```

```python
#4.find the days inwhich temperature greaterthan 30 or windspeed lessthan 4?
df['day'][(df['temperature']>30) | (df['windspeed']<4)]
```

```python
#5.to print event and day where temperature lessthan 20 or windspeed greaterthan 5
df[['day','event']][(df['temperature']<20) | (df['windspeed']>5)]
```

```python
#6.to print event and day where temperature lessthan 20 or windspeed greaterthan 5 and inwhich temperature greaterthan 30 or windspeed lessthan 4?
df[['day','event']][((df['temperature']<20) | (df['windspeed']>5)) & ((df['temperature']>30) | (df['windspeed']<4))]
```

```python
#7.find the days, temperature and event where temperature >=35 and windspeed>=5

df[['day','temperature','event']][(df['temperature']>=35) & (df['windspeed']>=5)]
```

```python
#8.to print event where temperature <=28 and windspeed <=5

df['event'][(df['temperature']<=28) & (df['windspeed']<=5)]
```

```python
#9.find the days, temperature and event where temperature >=35 and windspeed>=5 or where temperature <=28 and windspeed <=5

df[['day','temperature','event']][((df['temperature']>=35) & (df['windspeed']>=5)) | ((df['temperature']<=28) & (df['windspeed']<=5))]
```
