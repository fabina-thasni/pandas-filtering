**PANDAS SERIES**

Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index.

Pandas Series A Series is a one-dimensional array-like object that can hold various types of data, such as integers, floats, strings, and even other Python objects. It is similar to a column in an Excel spreadsheet or a single column in a database table.

Key Characteristics of a Series:

One-Dimensional: It holds a single column of data. Indexed: Each element in a Series has an associated index, which can be numeric (default) or a custom index. Homogeneous: All elements in a Series are of the same data type.

**DATA FRAMES**

DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object

Along with the data, you can optionally pass index (row labels) and columns (column labels) arguments. If you pass an index and / or columns, you are guaranteeing the index and / or columns of the resulting DataFrame. Thus, a dict of Series plus a specific index will discard all data not matching up to the passed index.

If axis labels are not passed, they will be constructed from the input data based on common sense rules.

```python
import pandas as pd
import numpy as np
path="/content/weather_data.csv"
df=pd.read_csv(path)
df
```

```python
#1.find dataframe where windspeed greater than 5 and event is equal to snow
df[(df['windspeed']>5) & (df['event']=='Snow')]
```

```python
#2.find dataframe where windspeed<5 and event not equal to rain?
df[(df['windspeed']<5) & (df['event']!='Rain')]
```

```python
#3.find dataframe where windspeed greater than 5 and event is equal to snow or dataframe where windspeed<5 and event not equal to rain?

df[((df['windspeed']>5) & (df['event']=='Snow')) | ((df['windspeed']<5) & (df['event']!='Rain'))]
```

```python
#4.find the days inwhich temperature greaterthan 30 or windspeed lessthan 4?
df['day'][(df['temperature']>30) | (df['windspeed']<4)]
```

```python
#5.to print event and day where temperature lessthan 20 or windspeed greaterthan 5
df[['day','event']][(df['temperature']<20) | (df['windspeed']>5)]
```

```python
#6.to print event and day where temperature lessthan 20 or windspeed greaterthan 5 and inwhich temperature greaterthan 30 or windspeed lessthan 4?
df[['day','event']][((df['temperature']<20) | (df['windspeed']>5)) & ((df['temperature']>30) | (df['windspeed']<4))]
```

```python
#7.find the days, temperature and event where temperature >=35 and windspeed>=5

df[['day','temperature','event']][(df['temperature']>=35) & (df['windspeed']>=5)]
```

```python
#8.to print event where temperature <=28 and windspeed <=5

df['event'][(df['temperature']<=28) & (df['windspeed']<=5)]
```

```python
#9.find the days, temperature and event where temperature >=35 and windspeed>=5 or where temperature <=28 and windspeed <=5

df[['day','temperature','event']][((df['temperature']>=35) & (df['windspeed']>=5)) | ((df['temperature']<=28) & (df['windspeed']<=5))]
```
